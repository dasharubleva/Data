1. Запишіть параметри задачі у змінні Python:
    - `sample_mean` - $\bar{x}$, середнє значення в вибірці
    - `population_mean` - $\mu_0$, середнє значення в популяції (тобто те, з яким ми порівнюємо середнє вибіркове значення)
    - `population_std` - $\sigma$, вибіркове стандартне відхилення, яке ми вважаємо, що є рівним ст. відх. популяції, адже маємо достатньо велику вибірку.

```python
sample_mean =  np.mean(data)
population_mean = 20
population_std = np.std(data, ddof=1)
sample_size = len(data)
alpha = 0.05
```

ГІПОТЕЗА:
============================================================


H₀: μ = 20 ($ продажі в середньому на чек не виросли)

H₁: μ > 20 ($ продажі в сердньому на чек виросли за липень)

Тип тесту: правосторонній (one-tailed)

Чому саме правосторонній тест?

Ми намагаємося з'ясувати, чи наші продажі за липень виросли , порівняно з середніми в попередні місяця, тому маємо перевірити чи чек вищий, а не просто вдрізняється.


ОБЧИСЛЕННЯ СТАНДАРТНОЇ ПОМИЛКИ:
============================================================
Формула: SE = σ / √n

SE = 6.25356123240434 / √100

SE = 6.25356123240434 / 10

SE = 0.625

Інтерпретація: В середньому, вибіркове середнє відхиляється
від справжнього середнього на ±0.625356123240434 $ за середній чек.

```python
standard_error = population_std / np.sqrt(sample_size)
print(standard_error)
print(population_std)
print(len(data))
print(sample_mean)
```

РОЗРАХУНОК Z-СТАТИСТИКИ:
============================================================
Формула: z = (x̄ - μ₀) / SE

z = (19.37755319670441 - 20) / 0.625

z = -0.6224468033 / 0.625

z = -0.995347738933505 = близько - 0.995

Інтерпретація:
Наше вибіркове середнє відхиляється на - 0.995 стандартних
помилок від гіпотетичного середнього популяції.


```python
z_statistic = (sample_mean - population_mean) / standard_error
z_statistic
```

============================================================

КРИТИЧНЕ ЗНАЧЕННЯ:

============================================================

Для рівня значущості α = 0.05 (правосторонній тест)

Критичне z-значення = близько 1.644



```python
z_critical = stats.norm.ppf(1 - alpha)
z_critical
```

============================================================

P-VALUE:

============================================================

p-value = 0.8402164165464012

Що це означає?
Якщо H₀ правильна (сердній чек не виріс), то ймовірність
випадково отримати середній чек  ≥ 19.37 становить 0.8402164165464012
або приблизно 0.84 %

Це НАДЗВИЧАЙНО мало! (менше 1%)

```python
p_value = 1 - stats.norm.cdf(z_statistic)
p_value
```

============================================================

Порівняння z-статистики з критичним значенням

============================================================

z-статистика = -0.995

Критичне значення = 1.644

Порівняння: -0.995 < 1.644

Висновок: z-статистика менша за критичне значення

Рішення: не ВІДХИЛЯЄМО H₀

============================================================

Порівняння p-value з рівнем значущості

============================================================


p-value = 0.840

Рівень значущості α = 0.05

Порівняння: 0.840 > 0.05

Висновок: p-value більше за рівень значущості

Рішення: не ВІДХИЛЯЄМО H₀

============================================================

ФІНАЛЬНИЙ ВИСНОВОК:

============================================================

Чи зійшлись значення z-статистику та р-значення в цьому завданні з попередніми обчисленнями?

```python
from statsmodels.stats import weightstats as stests
```

```python
z_stat_sm, p_val_sm = stests.ztest(data,
                                  value=population_mean,
                                  alternative='larger')

print(f"Результати z-тесту:")
print(f"  Z-статистика: {z_stat_sm:.3f}")
print(f"  P-value: {p_val_sm:.6f}")
print()
```

ІНТЕРПРЕТАЦІЯ:
----------------------------------------
   НЕ ВІДХИЛЯЄМО H₀

   p-value (0.840) > α (0.05)

   Висновок: при рівні значущості 0.05, ми можемо стверджувати , що нам не вдалося підняти середній чек за липень.
